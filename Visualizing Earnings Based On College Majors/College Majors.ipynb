# Visualizing Earnings Based On College Majors



## 1. Introduction



In this project, I explored how using the pandas and matplotlib plotting functionality along with the Jupyter notebook interface allows me to explore data quickly using visualizations.

I worked with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. FiveThirtyEight cleaned the dataset and released it on their Github repo.

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:


- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.


```python
#Import pandas and matplotlib into the environment
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

Read the dataset into a DataFrame and start exploring the data.

```python
recent_grads = pd.read_csv("recent-grads.csv")
```

## 2. Scatter Plots

```python
#Sample_size and Median
recent_grads.plot(kind ='scatter', x ='Sample_size', y = 'Median')

plt.title("Relation between Sample_size & Median")
plt.xlabel("Sample_size")
plt.ylabel("Median")

plt.show()
```

```python
#Sample_size and Unemployment_rate
recent_grads.plot(kind ='scatter', x ='Sample_size', y = 'Unemployment_rate')

plt.title("Relation between Sample_size & Unemployment_rate")
plt.xlabel("Sample_size")
plt.ylabel("Unemployment_rate")

plt.show()
```

```python
#Full_time and Median
recent_grads.plot(kind ='scatter', x ='Full_time', y = 'Median')

plt.title("Relation between Full_time & Median")
plt.xlabel("Full_time")
plt.ylabel("Median")

plt.show()
```

```python
#ShareWomen and Unemployment_rate
recent_grads.plot(kind ='scatter', x ='ShareWomen', y = 'Unemployment_rate')

plt.title("Relation between ShareWomen & Unemployment_rate")
plt.xlabel("ShareWomen")
plt.ylabel("Unemployment_rate")

plt.show()
```

```python
#Men and Median
recent_grads.plot(kind ='scatter', x ='Men', y = 'Median')

plt.title("Relation between Men & Median")
plt.xlabel("Men")
plt.ylabel("Median")

plt.show()
```

```python
#Women and Median
recent_grads.plot(kind ='scatter', x ='Women', y = 'Median')

plt.title("Relation between Women & Median")
plt.xlabel("Women")
plt.ylabel("Median")

plt.show()
```

## 3. Histograms

```python
cols = ["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate","Men","Women"]

for c in cols:
    print("")
    count, bin_edges = np.histogram(recent_grads[c])
    recent_grads[c].plot(kind = 'hist',xticks= bin_edges, figsize=(10,8))
    plt.title(c)
    plt.xlabel(c)
    plt.show()
```
